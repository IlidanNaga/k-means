```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import math

def leng(x,y):
    
    z = x - y
    z **= 2
    res = z.sum()    
    return math.sqrt(res)

clusters_am = 4
centers_am = 2
amount = 1500
rang = 10
precise = 2
directions = 2
distraction = 0


centers = (np.empty(centers_am * directions)).reshape(-1, directions)

for i in range(centers_am):
    
    for j in range(directions):
        
        centers[i][j] = np.round(np.random.uniform(-1 * rang, rang), precise)

centers
```

```python
len = (np.empty(centers_am ** 2)).reshape(centers_am, -1)

for i in range(centers_am):
    
    for j in range(centers_am):
        
        len[i][j] = leng(centers[i],centers[j])
        
    len[i][i] = 4 * rang
        
len = np.round(len, precise)

circles = len.min(axis = 0) / 2

circles
```

```python
in_preset = int(amount / (centers_am + distraction))
outside_preset = amount - in_preset * centers_am

preset = np.round((1 / math.sqrt(2)) * np.random.rand(in_preset * centers_am * directions), precise)
preset = preset.reshape(-1, directions)

for i in range(in_preset):
    
    for j in range(centers_am):
        
        preset[j*in_preset + i] *= circles[j]
        preset[j*in_preset + i] += centers[j]

        for k in range(directions):
            
            while (preset[j*in_preset + i][k] > 10):
                preset[j*in_preset + i][k] -= 1
            
            while (preset[j*in_preset +i][k] < -10):
                preset[j*in_preset + i][k] += 1 
        
others = (np.round(20 * (0.5 - np.random.rand(outside_preset * directions)), precise))
others = others.reshape(-1, directions)

total = np.concatenate([preset,others])
```

```python
#attempt to generate once again

if distraction == 0:
    in_preset = amount
    out_preset = 0
else:
    in_preset = int(amount / (centers_am + distraction)) * centers_am
    out_preset = amount - in_preset

preset_points = (math.sqrt(2)* (0.5 - np.random.rand(in_preset * directions))).reshape(-1, directions)
sent_to = np.random.randint(0, centers_am, in_preset)

for i in range(in_preset):
    
    preset_points[i] *= circles[sent_to[i]]
    preset_points[i] += centers[sent_to[i]]
    
    for k in range(directions):
        
        while preset_points[i][k] > 10:
            preset_points[i][k] -= 1
        
        while preset_points[i][k] < -10:
            preset_points[i][k] += 1
            
other_points = (20 * (0.5 - np.random.rand(out_preset))).reshape(-1, directions)

total_points = np.round(np.concatenate([preset_points, other_points]), precise)

points = np.hstack([total_points, (np.zeros(amount)).reshape(amount, -1)])
```

```python
fig = plt.figure()

for i in range(amount):
    
    plt.scatter(points[i][0],points[i][1],color='b', edgecolors = 'w')
    
for i in range(centers_am):
    
    plt.scatter(centers[i][0],centers[i][1],color = 'r', edgecolors = 'w')
    
```

```python
clusters_cent = (np.round(20 * (0.5 - np.random.rand(clusters_am * directions)), precise)).reshape(-1, directions)

def check_range(x,y):
    k = np.concatenate([y, np.zeros(1)])
    z = x - k
    z[directions] = 0
    z **= 2
    return math.sqrt(z.sum())


flag=True
    
while flag:

    fig=plt.figure()
    
    plt.xlabel('first measure')
    plt.ylabel('second measure')
    
    for i in range(amount):
        m_range=2*rang;
        for j in range(clusters_am):

            curr=check_range(points[i],clusters_cent[j])

            if curr<=m_range :
                m_range=curr
                points[i][directions]=j


    for i in range(amount):

        plt.scatter(points[i][0],points[i][1],color=colors[int(points[i][2])],edgecolors = 'w')

    new_centers=(np.empty(2*clusters_am)).reshape(-1,2)
    
    for j in range (clusters_am):
        
        curr_x=0;
        x_am=0;
        curr_y=0;
        y_am=0;
        
        for i in range (amount):
            
            if points[i][2]==j:
                
                curr_x+=points[i][0]
                curr_y+=points[i][1]
                
                x_am+=1
                y_am+=1
                
        new_centers[j][0]=curr_x/x_am
        new_centers[j][1]=curr_y/y_am
        
    new_flag=True
    
    for j in range(clusters_am):
        
        for k in range(2):
            
            if clusters_cent[j][k]!=new_centers[j][k] :
                clusters_cent[j][k]=new_centers[j][k]
                new_flag=False
            
    flag=not new_flag
    
    plt.show()
                
```
